 2. Looping Statements

Loops are used to execute a block of code repeatedly.

### 2.1 `for` Loops

`for` loops are used to iterate over a *sequence* (like a list, tuple, dictionary, set, or string) or other *iterable* object.

**Syntax:**

```python
for variable in iterable:
    # Code block to execute for each item in the iterable
    # 'variable' takes the value of the current item in each iteration
    print(variable)

while loops execute a block of code  a specified condition remains True.
while condition:
# Code block to execute as long as condition is True
print("Condition is still True")
# IMPORTANT: Usually, something inside the loop needs to
# change the condition, otherwise it might loop forever!

````python
count=0
print("Counting up to 5 using while loop:")
while count < 5:
    print(f"Count is: {count}")
    count += 1

print("Loop finished.")
print("-" * 20)
````

````python
count=10
while count > 5:
    print(f"Count is: {count}")
    count -= 1

print("Loop finished.")
print("-" * 20)
````

````python
count=20
while count < 50:
    print(f"Count is: {count}")
    count *= 2

print("Loop finished.")
print("-" * 20)
````

````python
user_input = ""
print("Enter 'quit' to exit")
while user_input != "QUIT":
    user_input = input("Enter command: ")
    print(f"You entered: {user_input}")
    user_input= user_input.upper()  #No matter what case you use will work and change it into upper case to match the QUIT

print("Exited the loop because input was 'quit'.")
````

## 3. Loop Control Statements (`break`, `continue`, `pass`)

These statements allow you to modify the standard behavior of loops.

*   **`break`**: â€” Stop the loop right now!
*   **`continue`**: Skip this one and move on.
*   **`pass`**: Does nothing. It acts as a placeholder where syntax requires a statement, but you don't want any code to execute. This is often used in empty function definitions, classes, or conditional blocks during development.

````python
# Example 1: break
print("Finding the first multiple of 7 between 1 and 20:")
lists = []
for number in range(1, 21):
   
    if number % 7 == 0:
        print(f"Found it! {number} is a multiple of 7.")
        break
    else:
        lists.append(number)

print("Loop finished or broken.")
print("Lists:", lists)
````
